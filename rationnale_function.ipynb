```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Plotter:
    def __init__(self):
        self.x = np.linspace(-10, 10, 1000)
        self.x = np.delete(self.x, np.where(self.x == 1))
        self.y = self.x - 1 + 3/(self.x -1)
        self.a_o = (self.x - 1)
        self.xox = np.full_like(self.x, 0)
        

    def plot(self):
        plt.figure(figsize=(6, 6))
        plt.axis('equal')
        plt.plot(self.x,self.a_o, label='Asymptote')
        plt.plot(self.x, self.xox, label='Horizontal Asymptote')
        plt.plot(self.x, self.y)
        plt.xlim(-10, 10)
        plt.ylim(-10, 10)
        plt.title("Rationnal Function Plot")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.grid(True, which='both', linestyle='--', linewidth=0.5)
        plt.show()
```

```python
p = Plotter()
p.plot()
# This code defines a class `Plotter` that creates a plot of a rational function.
```
