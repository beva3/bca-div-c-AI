```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Plotter:
    def __init__(self):
        self.__initialize_range()
        
    def __initialize_range(self):
        # analysis the define domain of the function
        self.x = np.linspace(-10, 10, 1000)
        self.x = np.delete(self.x, np.where(self.x == -1)) # remove the point where the function is not defined
    
    def __compute_function(self):
        # compute the function value
        self.y = self.x + 1 + 3/(self.x + 1)
        
        self.__av = self.x + 1
        
    
    def __setup_plot(self):
        # setup the plot
        plt.figure(figsize=(10, 10))
        plt.axis('equal')
        plt.title('rationnal of the function')
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.xlim(-10, 10)
        plt.ylim(-10, 10)
        plt.grid(color='gray',which='both', linestyle='--', linewidth=0.5)
        
    def plot(self):
        # full ploting method
        self.__compute_function()
        self.__setup_plot()
        plt.plot(self.x, self.y, label='f(x) = x - 1 + 3/(x + 1)', color='blue')
        plt.plot(self.x, self.__av, label='y = x + 1', color='red', linestyle='--')
        plt.axhline(0, color='black', lw=1)
        
```

```python
my_plot = Plotter()
my_plot.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

class Plotter:
    def __init__(self):
        self.__my_function()
    
    def __my_function(self):
        # define the function
        self.x = np.linspace(-10, 10, 1000)
        self.x = np.delete(self.x, np.where(self.x == -1))
        self.y = self.x + 1 + 3/(self.x + 1)
        
    def __setup_plot(self):
        # setup the plot
        plt.figure(figsize=(10, 10))
        plt.axis('equal')
        plt.title('Learning to plot')
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.xlim(-10, 10)
        plt.ylim(-10, 10)
        plt.grid(color='gray',which='both', linestyle='--', linewidth=0.5)
    
    def plot(self):
        # full ploting method
        self.__setup_plot()
        plt.plot(self.x, self.y, label='f(x) = x - 1 + 3/(x + 1)', color='blue')
        plt.plot(self.x, self.x + 1, label='y = x + 1', color='red', linestyle='--')
        plt.axhline(0, color='black', lw=1)
        plt.axvline(0, color='black', lw=1)
        plt.legend()
        plt.show()


# Example usage   
my_plot = Plotter()
my_plot.plot()
```
